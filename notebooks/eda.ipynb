```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv("../data/Final_Regression.csv")
```

### Scatter Plot of Budget vs Revenue

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(data=df, x='budget', y='revenue')
plt.title('Scatter Plot of Budget vs Revenue')
plt.xlabel('Budget')
plt.ylabel('Revenue')
plt.show()
```

### Histograms



```python
plt.figure(figsize=(12, 6))
plt.subplot(1, 2, 1)
sns.histplot(df['budget'], bins=20, kde=True)
plt.title('Histogram of Budget')
plt.xlabel('Budget')
plt.ylabel('Frequency')

plt.subplot(1, 2, 2)
sns.histplot(df['revenue'], bins=20, kde=True)
plt.title('Histogram of Revenue')
plt.xlabel('Revenue')
plt.ylabel('Frequency')
plt.tight_layout()
plt.show()
```

### Box Plot

```python
plt.figure(figsize=(10, 6))
sns.boxplot(x='budget', y='revenue', data=df)
plt.title('Box Plot of Revenue by Budget')
plt.xlabel('Budget')
plt.ylabel('Revenue')
plt.show()
```

### Correlation Matrix Heatmap

```python
plt.figure(figsize=(10, 8))
corr = df[['budget', 'revenue', 'release_year', 'Adventure', 'Romance', 'Action']].corr()
sns.heatmap(corr, annot=True, cmap='coolwarm', fmt=".2f", linewidths=0.5)
plt.title('Correlation Matrix Heatmap')
plt.show()
```

### Line Plot of Revenue over Time

```python
plt.figure(figsize=(10, 6))
sns.lineplot(data=df, x='release_year', y='revenue')
plt.title('Average Revenue Over Time')
plt.xlabel('Release Year')
plt.ylabel('Average Revenue')
plt.show()
```

### Bar Plot of Revenue by Genre

```python
genre_cols = ['Adventure', 'Romance', 'Action']
genre_revenue = df.groupby(genre_cols)['revenue'].mean().reset_index()
plt.figure(figsize=(10, 6))
sns.barplot(data=genre_revenue, x='Adventure', y='revenue', hue='Romance')
plt.title('Average Revenue by Genre')
plt.xlabel('Adventure')
plt.ylabel('Average Revenue')
plt.show()
```

```python
# Plot 1: Distribution of Movies by Genre
plt.figure(figsize=(10, 6))
genre_counts = df.iloc[:, 31:-1].sum().sort_values(ascending=False)
sns.barplot(x=genre_counts.values, y=genre_counts.index)
plt.title('Distribution of Movies by Genre')
plt.xlabel('Count')
plt.ylabel('Genre')
plt.show()

# Plot 2: Relationship Between Budget and Revenue
plt.figure(figsize=(8, 6))
sns.scatterplot(data=df, x='budget', y='revenue')
plt.title('Relationship Between Budget and Revenue')
plt.xlabel('Budget')
plt.ylabel('Revenue')
plt.show()

# Plot 3: Average Actor Age by Gender
plt.figure(figsize=(8, 6))
sns.boxplot(data=df, x='actor_0_gender', y='Average Actor Age')
plt.title('Average Actor Age by Gender')
plt.xlabel('Gender (0: Female, 1: Male)')
plt.ylabel('Average Actor Age')
plt.xticks(ticks=[0, 1], labels=['Female', 'Male'])
plt.show()

# Plot 4: Runtime Distribution
plt.figure(figsize=(8, 6))
sns.histplot(data=df, x='runtime', bins=20, kde=True)
plt.title('Runtime Distribution')
plt.xlabel('Runtime')
plt.ylabel('Frequency')
plt.show()

# Plot 5: Production Companies Revenue Comparison
plt.figure(figsize=(10, 6))
revenue_by_company = df.groupby('production_company_name')['revenue'].sum().sort_values(ascending=False)
sns.barplot(y=revenue_by_company.index, x=revenue_by_company.values)
plt.title('Production Companies Revenue Comparison')
plt.xlabel('Total Revenue')
plt.ylabel('Production Company')
plt.show()

# Plot 6: Release Month vs. Revenue
plt.figure(figsize=(10, 6))
sns.lineplot(data=df, x='release_month', y='revenue', estimator='mean')
plt.title('Release Month vs. Revenue')
plt.xlabel('Release Month')
plt.ylabel('Revenue')
plt.show()

# Plot 7: Gender Ratio of Lead Actors
plt.figure(figsize=(6, 6))
gender_ratio = df['actor_0_gender'].value_counts(normalize=True)
gender_ratio.plot(kind='pie', autopct='%1.1f%%')
plt.title('Gender Ratio of Lead Actors')
plt.ylabel('')
plt.legend(['Male', 'Female'])
plt.show()

# Plot 8: Average Revenue by Genre
plt.figure(figsize=(10, 6))
avg_revenue_by_genre = df.iloc[:, 31:-1].multiply(df['revenue'], axis=0).mean().sort_values(ascending=False)
sns.barplot(x=avg_revenue_by_genre.values, y=avg_revenue_by_genre.index)
plt.title('Average Revenue by Genre')
plt.xlabel('Average Revenue')
plt.ylabel('Genre')
plt.show()

# Plot 9: Actor Popularity Distribution
plt.figure(figsize=(8, 6))
sns.histplot(data=df, x='actor_popularity', bins=20, kde=True)
plt.title('Actor Popularity Distribution')
plt.xlabel('Actor Popularity')
plt.ylabel('Frequency')
plt.show()

# Plot 10: Age Distribution of Actors
plt.figure(figsize=(8, 6))
sns.histplot(data=df.iloc[:, 26:31].stack().str.extract('(\d{4})').astype(float).mean(axis=1), bins=20, kde=True)
plt.title('Age Distribution of Actors')
plt.xlabel('Actor Age')
plt.ylabel('Frequency')
plt.show()
```

```python
# Group by production company and sum the revenue
revenue_by_company = df.groupby('production_company_name')['revenue'].sum()

# Get the top 10 production companies by revenue
top_10_companies = revenue_by_company.nlargest(10)

print(top_10_companies)

plt.figure(figsize=(10, 6))
top_10_companies.plot(kind='bar', color='skyblue')
plt.title('Top 10 Production Companies by Total Revenue')
plt.xlabel('Production Company')
plt.ylabel('Total Revenue (in billions)')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

```python
# Get the top 10 production companies
top_10_production_companies = top_10_companies.index.tolist()

# Dictionary to store top 10 movies for each production company
top_10_movies_by_company = {}

# Iterate through each production company
for company in top_10_production_companies:
    # Filter the dataframe for movies produced by the current production company
    company_movies = df[df['production_company_name'] == company]
    # Sort the movies by revenue in descending order and get the top 10
    top_10_movies = company_movies.sort_values(by='revenue', ascending=False).head(10)
    # Store the top 10 movies for the current production company
    top_10_movies_by_company[company] = top_10_movies[['overview', 'revenue']]

# Print out the top 10 movies for each production company
for company, movies in top_10_movies_by_company.items():
    print(f"\nTop 10 Movies for {company}:")
    for index, movie in movies.iterrows():
        print(f"{movie['overview']} - Revenue: ${movie['revenue']:.2f} million")
```

### Genre

```python
import pandas as pd
import matplotlib.pyplot as plt

# Assuming df is the DataFrame containing the movie data

# Define a list of genres
genres = ['Crime', 'Drama', 'War', 'Western', 'Family', 'Thriller', 'Fantasy', 
          'Mystery', 'Animation', 'Music', 'Romance', 'Adventure', 'History', 
          'Science Fiction', 'Comedy', 'TV Movie', 'Documentary', 'Horror', 'Action']

# Initialize a dictionary to store total revenue for each genre
genre_revenue = {genre: 0 for genre in genres}

# Iterate over each movie and update the genre revenue
for genre in genres:
    genre_revenue[genre] = df[df[genre] == 1]['revenue'].sum()

# Convert the dictionary to a pandas Series for easy plotting
genre_revenue_series = pd.Series(genre_revenue)

# Sort the genre revenue in descending order
sorted_genre_revenue = genre_revenue_series.sort_values(ascending=False)

# Plotting the revenue by genre
plt.figure(figsize=(10, 6))
sorted_genre_revenue.plot(kind='bar', color='skyblue')
plt.title('Revenue by Genre')
plt.xlabel('Genre')
plt.ylabel('Total Revenue (in $ billions)')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```
